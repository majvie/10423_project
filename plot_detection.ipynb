```python
# read csv from pandas
import matplotlib.pyplot as plt
import pandas as pd
import re
import os


# Unbiased
df = pd.read_csv('bias_detection_results_acc_none.csv')
grouped_df_unbiased = df.groupby('number_of_reviews')['accuracy_rate'].mean().reset_index()
grouped_df_unbiased = grouped_df_unbiased.sort_values('number_of_reviews')
x_unbiased = grouped_df_unbiased["number_of_reviews"].tolist()
y_unbiased = grouped_df_unbiased["accuracy_rate"].tolist()

# Ranking
df = pd.read_csv('bias_detection_results_acc_ranking.csv')
grouped_df_rank = df.groupby('number_of_reviews')['accuracy_rate'].mean().reset_index()
groupegrouped_df_rankd_df = grouped_df_rank.sort_values('number_of_reviews')
x_rank = grouped_df_rank["number_of_reviews"].tolist()
y_rank = grouped_df_rank["accuracy_rate"].tolist()

# Prompt

df = pd.read_csv('bias_detection_results_acc_prompt.csv')
grouped_df_prompt = df.groupby('number_of_reviews')['accuracy_rate'].mean().reset_index()
grouped_df_prompt = grouped_df_prompt.sort_values('number_of_reviews')
x_prompt = grouped_df_prompt["number_of_reviews"].tolist()
y_prompt = grouped_df_prompt["accuracy_rate"].tolist()

# Filter
df = pd.read_csv('bias_detection_results_acc_filter.csv')
grouped_df_filter = df.groupby('number_of_reviews')['accuracy_rate'].mean().reset_index()
grouped_df_filter = grouped_df_filter.sort_values('number_of_reviews')
x_filter = grouped_df_filter["number_of_reviews"].tolist()
y_filter = grouped_df_filter["accuracy_rate"].tolist()


# 12 products, each datapoint per product 2 reviews -> each dot is the average of 24 estimations
plt.figure(figsize=(10, 6))
plt.plot(x_unbiased, y_unbiased, marker='o', linestyle='-', label='Unbiased')
plt.plot(x_rank, y_rank, marker='o', linestyle='-', label='Ranking')
plt.plot(x_prompt, y_prompt, marker='o', linestyle='-', label='Prompt')
plt.plot(x_filter, y_filter, marker='o', linestyle='-', label='Filter')
plt.title('Detecting the Fidelity of AI Summaries of Product Reviews')
plt.xlabel('Number of Reviews given to the Model')
plt.ylabel('Likert Scale (0 = Very Unfaithful, 10 = Very Faithful)')
plt.grid(axis='y', alpha=0.7)
#plt.ylim(3.0, 10.0)
plt.legend()
plt.show()
```
